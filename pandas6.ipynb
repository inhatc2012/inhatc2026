```python
df=pd.DataFrame(
    {"key1":list("aabbab"),
     "key2":["one","two","three"]*2,
     "data1":np.random.randn(6),
     "data2":np.random.randn(6)})
df
```

```python
df["data1"]
```

```python
group1=df["data1"].groupby(df["key1"])

# key1 컬럼의 값을 기준으로         ←-------- 기준: 컬럼
# data1 컬럼을                    ←-------- 대상: Series
# 그룹으로 나누어라
```

```python
df[["key1", "data1"]]
```

```python
# group1=df["data1"].groupby(df["key1"])에서 group은

# "기준이 무엇인 시리즈" ←----- 그룹 1
# "기준이 무엇인 시리즈" ←----- 그룹 2
# ...
# "기준이 무엇인 시리즈" ←----- 그룹 n
# 로 구성

# group1
# ├── 'a'             ←----- 그룹명
# │    0   -0.2816
# │    1   -0.3085
# │    4   -2.6236
# │
# └── 'b'             ←----- 그룹명
#      2    1.6443
#      3   -0.6814
#      5   -0.5932

group1
```

```python
# 1. 그룹 목록 확인
group1.groups

# {'a': [0, 1, 4], 'b': [2, 3, 5]}
# 키: 그룹명
# 값: 해당 그룹의 인덱스 리스트
```

```python
# 2. 각 그룹 순회하며 내용 출력

# group
# ├── 'a'             ←----- 그룹명 (name)
# │    0   -0.2816
# │    1   -0.3085
# │    4   -2.6236
# │
# └── 'b'             ←----- 그룹명 (name)
#      2    1.6443
#      3   -0.6814
#      5   -0.5932

for name, _ in group1:
    print(name)

print('-'*100)

for _, g in group1:
    print(g)

print('-'*100)

for name, g in group1:
    print(name, g)
```

```python
# 3. 특정 그룹
group1.get_group("a")
```

```python
group1.mean()
```

```python
group2=df["data1"].groupby([df["key1"], df["key2"]])
# key1 컬럼, key1 컬럼의 값을 기준으로         ←-------- 기준: 컬럼
# data1 컬럼을                              ←-------- 대상: Series
# 그룹으로 나누어라 → key1, key2가 모두 같아야 같은 그룹
```

```python
df[["key1", "key2", "data1"]]
```

```python
for name, _ in group2:
    print(name)

print('-'*100)

for _, g in group2:
    print(g)

print('-'*100)

for name, g in group2:
    print(name, g)
```

```python
ave = group2.mean()
ave
```

```python
ave.index

# MultiIndex([('a',   'one'),
#             ('a',   'two'),
#             ('b',   'one'),
#             ('b', 'three')],
#            names=['key1', 'key2'])

# 레벨 0: key1 ('a', 'b')  ← 바깥쪽
# 레벨 1: key2 ('one', 'two', 'three') ← 안쪽
```

```python
# unstack()은 인덱스를 컬럼으로 이동한다는 점에서 reset_index()와 유사
# 하지만 unstack()은 멀티인덱스 상황에서 특정 인덱스를 컬럼으로 이동

# unstack()의 디폴트는 가장 마지막 레벨을 컬럼으로
# 즉, 디폴트는 unstack(-1)

ave.unstack()
# 결측치가 생기는 이유는 {a, b} x {one, two , three} 모든 조합이 있지 않기 때문
```

```python
ave.unstack(0)
```

```python
df3 = pd.DataFrame({
    "팀": ["A", "A", "B", "B", "A", "B"],
    "점수": [85, 90, 78, 92, 88, 76],
    "출석": [20, 18, 22, 19, 21, 20]
})

group4 = df3.groupby("팀")
group4.mean()
```

```python
# 그룹 내의 여러 값을 하나로 줄여주는 함수를 집계 함수라고 함
# [0, 3]  →   sum(): 3
# [0, 3]  →  mean(): 1.5
# [0, 3]  →   min(): 0
# [0, 3]  →   max(): 3

# 기본 통계
# 'sum'     # 합계
# 'mean'    # 평균
# 'median'  # 중앙값
# 'min'     # 최솟값
# 'max'     # 최댓값
# 'std'     # 표준편차
# 'var'     # 분산
# 'count'   # 개수 (NaN 제외)
# 'size'    # 개수 (NaN 포함)

# 기타
# 'first'   # 첫 번째 값
# 'last'    # 마지막 값
# 'nunique' # 고유값 개수
```

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({
    'key': ['A', 'B', 'C', 'A', 'B', 'C'],
    'data1': range(6),
    'data2': rng.randint(0, 10, 6)
})
df
```

```python
# aggregate()는 여러 집계 함수를 한번에 적용
df.groupby('key').aggregate(['min', np.median, 'max'])
```

```python
df
```

```python
# groupby('key')는 컬럼 'key'의 unique 값을 인덱스로

df.groupby('key').std()
```

```python
df.groupby('key').filter(lambda x: x['data2'].std()>4)

# filter(조건)의 형태로 사용

# lambda x: 에서 x는 그룹에 해당; 그룹의 타입은 데이터프레임
# lambda x: x['data']는 그룹의 'data' 컬럼에 해당

# "data2 컬럼의 표준편차가 4보다 큰 그룹만 남겨라"
```

```python
df.groupby('key').filter(lambda x: len(x) == 2)        # 그룹 행 개수가 2 초과
```

```python
df.groupby('key').filter(lambda x: x['data2'].max() > 8)
```
